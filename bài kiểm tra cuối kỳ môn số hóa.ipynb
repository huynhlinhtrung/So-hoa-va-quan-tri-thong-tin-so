```python
#Câu 2:
# Tính điểm trung bình theo từng môn học
mean_scores = df.groupby("Course")["Score"].mean()

# Điền giá trị thiếu bằng điểm trung bình của môn tương ứng
df["Score"] = df.apply(lambda row: mean_scores[row["Course"]] if pd.isnull(row["Score"]) else row["Score"], axis=1)
```

```python
# Tính trung bình theo từng môn học trong từng Campus
mean_scores_by_campus = df.groupby(["Course", "Campus"])["Score"].mean()

# Điền giá trị thiếu dựa trên Course và Campus tương ứng
df["Score"] = df.apply(
    lambda row: mean_scores_by_campus[row["Course"], row["Campus"]] if pd.isnull(row["Score"]) else row["Score"], 
    axis=1
)
```

```python
#Câu 5:
import seaborn as sns
import matplotlib.pyplot as plt

# Tạo cột kết hợp Course_Gender để phân tích
df['Course_Gender'] = df['Course'] + "_" + df['Gender'].fillna('Unknown')

# Vẽ boxplot
plt.figure(figsize=(12, 6))
sns.boxplot(
    data=df,
    x='Course_Gender',
    y='Score',
    palette='Set2',
    showmeans=True,  # Hiển thị giá trị trung bình
    meanprops={'marker':'o', 'markerfacecolor':'white', 'markeredgecolor':'black'}
)
plt.title('Phân phối điểm theo Môn học và Giới tính', fontsize=14)
plt.xlabel('Nhóm (Môn_Giới tính)')
plt.ylabel('Điểm số')
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
# Tính toán IQR cho từng nhóm
outliers = {}
for group in df['Course_Gender'].unique():
    group_data = df[df['Course_Gender'] == group]['Score'].dropna()
    Q1 = group_data.quantile(0.25)
    Q3 = group_data.quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    outliers[group] = df[(df['Course_Gender'] == group) & ((df['Score'] < lower_bound) | (df['Score'] > upper_bound))]

# In kết quả outliers
for group, outlier_df in outliers.items():
    if not outlier_df.empty:
        print(f"Outliers trong nhóm {group}:")
        print(outlier_df[['Student_ID', 'Score']])
```

Đề xuất kế hoạch hỗ trợ sinh viên outliers
a. Sinh viên điểm thấp (dưới Q1 - 1.5*IQR):

- Nguyên nhân: Thiếu kiến thức nền tảng, phương pháp học không phù hợp.

- Giải pháp:

    Học kèm 1-1 với trợ giảng.

    Cung cấp tài liệu bổ trợ theo từng môn (ví dụ: sổ tay công thức Toán, video thí nghiệm Vật lý).

b. Sinh viên điểm cao bất thường (trên Q3 + 1.5*IQR):

- Nguyên nhân: Có thể gian lận hoặc năng lực vượt trội.

- Giải pháp:

    Kiểm tra lại bài thi để loại trừ gian lận.

    Mời tham gia nhóm nghiên cứu hoặc thi học sinh giỏi.

Giải thích tác động
- Công bằng giáo dục: Phát hiện sớm sinh viên yếu để can thiệp kịp thời.

- Tối ưu nguồn lực: Tập trung hỗ trợ vào các nhóm cần thiết (ví dụ: Nữ học Vật lý điểm thấp, Nam học Hóa điểm cao bất thường).

```python
#Câu 8:
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
sns.scatterplot(
    data=df,
    x='Attendance (%)',
    y='Score',
    hue='Campus',
    palette=['#FF6B6B', '#4ECDC4', '#45B7D1'],  # Màu riêng cho từng campus
    style='Course',  # Phân biệt môn học bằng marker
    s=100  # Kích thước điểm
)
plt.title('Mối quan hệ giữa Điểm danh và Điểm số theo Campus', fontsize=14)
plt.xlabel('Tỷ lệ điểm danh (%)')
plt.ylabel('Điểm số')
plt.grid(alpha=0.3)
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')  # Đặt legend bên ngoài
plt.show()
```

Tác động đến môi trường học tập
    Campus A: Tăng tính kỷ luật, giảm tình trạng bỏ học.

    Campus B: Giảm áp lực vô ích, tập trung vào chất lượng.

    Campus C: Cân bằng giữa tự chủ và kiểm soát.

Triển khai thực tế: 
    Bước 1: Thí điểm chính sách trong 1 học kỳ.

    Bước 2: Đo lường sự thay đổi điểm số trung bình.

    Bước 3: Điều chỉnh linh hoạt dựa trên phản hồi sinh viên.

```python
#Câu 9:
avg_scores = df.groupby('Course')['Score'].mean().sort_values(ascending=False)
print("Điểm trung bình theo môn học:")
print(avg_scores.round(2))
```

```python
#Vẽ biểu đồ:
plt.figure(figsize=(8,5))
avg_scores.plot(kind='bar', color=['#4CAF50', '#2196F3', '#FF9800'], edgecolor='black')
plt.title('Điểm Trung Bình Theo Môn Học', pad=20, fontsize=14)
plt.xlabel('Môn học')
plt.ylabel('Điểm trung bình')
plt.ylim(70, 90)
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Thêm giá trị số trên mỗi cột
for i, v in enumerate(avg_scores):
    plt.text(i, v + 0.5, str(round(v,1)), ha='center', fontweight='bold')
plt.show()
```

```python
#Câu 11:
# Bước 1: Loại bỏ các hàng có Attendance hoặc Score bị thiếu
df_clean = df.dropna(subset=["Attendance", "Score"])

# Bước 2: Nhóm theo Course và tính trung bình Attendance và Score
course_stats = df_clean.groupby("Course")[["Attendance", "Score"]].mean().reset_index()

# Bước 3: Làm tròn cho đẹp
course_stats = course_stats.round(2)

# In kết quả
print(course_stats)
```

```python
#Câu 12:
from scipy.stats import skew

# Bỏ NaN trước khi tính skewness
score_data = df["Score"].dropna()

# Tính skewness
score_skewness = skew(score_data)

# In kết quả
print("Skewness của cột Score là:", round(score_skewness, 3))
```

Đề xuất điều chỉnh điểm & phân tích tác động:
Nếu skewness > 0 (lệch phải):
Đề xuất: Áp dụng chuẩn hóa điểm (z-score) hoặc phân phối theo phần trăm (percentile) để giảm thiểu ảnh hưởng của điểm cao bất thường.

Giải thích:
Khi có skewness dương, phần lớn sinh viên có điểm thấp hơn trung bình.

Việc chuẩn hóa giúp đưa các điểm số về thang điểm cân đối hơn → phản ánh đúng thực lực hơn thay vì bị điểm cao kéo lệch.

```python
#Câu 13:
import seaborn as sns
import matplotlib.pyplot as plt

# Lựa chọn các biến quan trọng
analysis_vars = ['Score', 'Attendance (%)', 'Study_Hours', 'Efficiency']

# Vẽ pairplot với regression line
sns.pairplot(
    df[analysis_vars].dropna(),
    kind='reg',
    plot_kws={'line_kws':{'color':'red'}, 'scatter_kws': {'alpha':0.5}},
    diag_kind='kde'
)
plt.suptitle('Mối quan hệ đa biến giữa các yếu tố học tập', y=1.02)
plt.show()
```
